```python
import pandas as pd
from pycoingecko import CoinGeckoAPI
from datetime import datetime
import numpy as np
```

```python
cg = CoinGeckoAPI()
```

```python
data = cg.get_coin_market_chart_by_id(id = 'bitcoin', vs_currency = 'usd', days = 30, interval = 'daily')
```

```python
price = pd.DataFrame(data['prices'], columns=['Date','Price'])
price
```

```python
price['Date'] = pd.to_datetime(price['Date'], unit='ms')
```

```python
def get_price(ids, currency, day, intervals):
  data = cg.get_coin_market_chart_by_id(id=ids, vs_currency=currency, days=day, interval=intervals)
  price = pd.DataFrame(data['prices'], columns=['Date','Price'])
  price['Date'] = pd.to_datetime(price['Date'], unit='ms')

  return price
```

```python
df = get_price('bitcoin', 'usd', 180, 'daily')
```

```python
df
```

```python
def get_ex(n):
  exchange = cg.get_exchanges_list(per_page=n)
  ex = pd.DataFrame(exchange)
  ex = ex[['id','name','country','url','trust_score_rank']]

  return ex
```
