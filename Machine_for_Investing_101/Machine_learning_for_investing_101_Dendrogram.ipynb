```python
import pandas as pd
import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
data = yf.download("SPY TDEX.BK  GC=F BTC-USD ETH-USD DOGE-USD ADA-USD",
                   start="2020-01-01",end="2022-01-01")
```

```python
data
```

```python
df=data.Close
```

```python
df.pct_change().corr()
```

```python
#df.resample("B").last().tail(10)
```

```python
businessday= df.resample("B").last().iloc[1:]
```

```python
businessday = businessday.fillna(method='ffill')
```

```python
len(businessday)
```

```python
ret = businessday.pct_change().iloc[1:]
```

```python
ret
```

```python
corr = ret.corr()
```

```python
corr
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage

linked = linkage(corr,'ward')

labelList = businessday.columns.to_list()

plt.figure()
dendrogram(linked,
         orientation= 'left',
         labels=labelList,
          leaf_rotation=0)
plt.show()
```
