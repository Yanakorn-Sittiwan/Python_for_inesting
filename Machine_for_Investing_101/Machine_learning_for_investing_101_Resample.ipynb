```python
import pandas as pd
import pandas_ta as ta
import numpy as np
import yfinance as yf
from datetime import datetime
```

```python
data = pd.DataFrame()
data = data.ta.ticker('AAPL', period = '3y', interval ='1d')
```

```python
df = data.copy()
```

```python
df.resample('W-Mon').last()
```

```python
df.tail()
```

### การทำ resample ที่ถูกต้องใช้ aggregate

### Week

```python
df_weekly = df.resample('W-Fri').agg({'Open': 'first',
                                     'High':'max',
                                     'Low':'min',
                                     'Close':'last',
                                     'Volume' : 'sum'})
```

```python
df_weekly.tail(10)
```

### Month

```python
df_monthly = df.resample('BM').agg({'Open': 'first',
                                     'High':'max',
                                     'Low':'min',
                                     'Close':'last',
                                     'Volume' : 'sum'})
```

```python
df_monthly
```

### Day

```python
df_everyday = df.resample('D').last()
```

```python
df_everyday
```

```python
df_businessday = df_everyday.resample('B').last()
df_businessday
```
