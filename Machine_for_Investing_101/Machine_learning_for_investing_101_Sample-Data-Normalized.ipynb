# Outlier&MinMaxScaler

```python
df_x_normalized = pd.DataFrame(X,columns=data.data.columns)
```

```python
df_x_normalized.describe()
```

```python
upper = df_x_normalized.quantile(.90)
upper
```

```python
low = .05
high = .95
quant_df = df_x_normalized.quantile([low, high])
```

```python
quant_df
```

```python
filt_df_x = df_x_normalized.apply(lambda x: x[(x>=quant_df.loc[low,x.name]) &  (x <= quant_df.loc[high,x.name])], axis=0)
```

```python
filt_df_x
```

```python
filt_df_x.dropna()
```

```python
Q1 = df_x_normalized.quantile(0.05)
Q3 = df_x_normalized.quantile(0.95)
```

```python
# อีกวิธีหนึ่ง
# ถ้าใส่ ~ เครื่องหมาย < จะเป็น >= ต้องระหวังเครื่องหมายให้ดี
df_x_normalized[~((df_x_normalized < Q1) | (df_x_normalized > Q3)).any(axis=1)]
```

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np
scaler = MinMaxScaler()
scaler.fit(df_x_normalized)
df_x_scaled = scaler.transform(df_x_normalized)
```
