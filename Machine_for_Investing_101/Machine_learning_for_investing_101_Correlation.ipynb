```python
import pandas as pd
import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
data = yf.download("SPY TDEX.BK AAPL GC=F BTC-USD ", start="2020-01-01",end="2022-01-01")
```

```python
data
```

```python
df = data.Close
```

```python
df
```

```python
pd.plotting.scatter_matrix(df.pct_change())
```

```python
pd.plotting.scatter_matrix(df.pct_change(),figsize=(10,10),marker='o',
                           hist_kwds={'bins':20} , alpha=0.8)
```

```python
sns.pairplot(df.pct_change(),diag_kws={'bins':10})
```

```python
df.corr()
```

```python
# ควรใช้ return ดู Correlation
df.pct_change().corr()
```

```python
corr = df.corr()
sns.heatmap(corr,
           xticklabels = corr.columns,
           yticklabels=corr.columns, annot=True)
```

```python
thai_stock = yf.download('PTT.Bk , PTTEP.Bk , KBANK.Bk , SCB.Bk , BBL.Bk , AOT.Bk  , RCL.Bk ,TTA.Bk ,PSL.Bk',
                        periods = "2y" , interval = '1d')
```

```python
thai_stock_close = thai_stock.Close
```

```python
thai_stock_close_corr = thai_stock_close.pct_change().corr()
```

```python
fig,ax= plt.subplots(figsize=(10,4))
sns.heatmap(thai_stock_close_corr,
           xticklabels = thai_stock_close_corr.columns,
           yticklabels=thai_stock_close_corr.columns, annot=True)
```

```python
# ต้อง fill nan ด้วย
df['SPY'].rolling(20).corr(df['GC=F'])
```

```python
df = df.fillna(method='ffill')
```

```python
df['SPY'].rolling(20).corr(df['GC=F'])
```

```python
df['SPY'].rolling(20).corr(df['GC=F']).plot()
```

```python
df['SPY'].pct_change().rolling(20).corr(df['GC=F'].pct_change()).plot()
```

```python
# ใช้ mean ดูเบื้องต้น
df['SPY'].pct_change().rolling(20).corr(df['GC=F'].pct_change()).mean()
```
