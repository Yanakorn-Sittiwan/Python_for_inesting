```python
import pandas as pd
import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt 
```

# หาความสัมพันธ์ด้วย scatterplot

```python
data = yf.download("SPY AAPL", start="2020-01-01",end="2022-01-01")
```

```python
data
```

```python
df=data.Close.copy()
```

```python
df.plot()
```

```python
df.pct_change(1).plot()
```

```python
df_ret = df.pct_change(1)
df_ret_weekly = df.pct_change(7)
```

```python
df_ret.plot.scatter(x="SPY", y='AAPL')
```

```python
df_ret_weekly.plot.scatter(x="SPY", y='AAPL')
```

```python
x = df_ret.SPY
y = df_ret.AAPL

x=x.dropna()
y=y.dropna()
```

```python
m,c = np.polyfit(x,y,1)
```

```python
plt.plot(x,y,'o')
plt.plot(x,m*x+c)
```
